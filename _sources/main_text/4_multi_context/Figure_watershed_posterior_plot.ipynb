### Posterior Plot
**Bar plot of distribution of Watershed posterior values for expression (eOutliers), proteomic (pOutliers) and splicing (sOutliers) outliers.**

#### Load libraries

```r
library(tidyverse)
library(showtext)
font_add_google("Noto Sans", "Noto Sans")
showtext_auto()
```

#### Load data

```r
full_posterior_filter <- read_rds("/restricted/projectnb/xqtl/jaempawi/xqtl/full_posterior_filter.rds")
```

#### Watershed posterior plot

```r
p1 <- ggplot(full_posterior_filter, aes(posterior, fill = ome)) +
    geom_histogram() +
    scale_y_log10() +
    geom_vline(xintercept = 0.5, color = "red") +
    facet_wrap(~ ome, ncol = 3) +
    theme_classic() +
    xlim(c(0,1)) +
    ylab("Subject-gene pairs") +
    xlab("Watershed posterior") +
    theme(text = element_text(family = "Noto Sans"),
        legend.position = "none",
        strip.background = element_blank(),
        strip.text = element_text(face = "bold", size = 20),
        axis.title.x = element_text(face = "bold", size = 24),
        axis.title.y = element_text(face = "bold", size = 24),
        axis.text.x = element_text(size = 20, color = "black"),
        axis.text.y = element_text(size = 20, color = "black"),
        panel.background = element_rect(fill = NA),
        panel.spacing = unit(1.5, "lines") 
    )
```

```r
options(repr.plot.width = 15, repr.plot.height = 6)
p1
```
