```r
# Load data and renamed metadata
Figure_1b <- readRDS("/restricted/projectnb/xqtl/jaempawi/xqtl/Staging/manuscript/xQTL_flagship_2025/Figure_1b_num_xQTL_loci_single_context.rds")
contexts_metadata <- read.csv("/restricted/projectnb/xqtl/jaempawi/xqtl/contexts_metadata_colors.csv", stringsAsFactors = FALSE)


# Preserve the original factor level order
original_order <- levels(Figure_1b$detectable$resource)
Figure_1b$detectable$resource <- as.character(Figure_1b$detectable$resource)

# Create mapping function to match data with metadata
apply_metadata_mapping <- function(resource_names, contexts_metadata) {
  result <- character(length(resource_names))
  
  for(i in seq_along(resource_names)) {
    name <- resource_names[i]
    
    cell_types <- c("OPC", "Mic", "Inh", "Ast", "Oli", "Exc")
    contains_cell <- any(sapply(cell_types, function(cell) grepl(cell, name)))
    
    if(contains_cell) {
      metadata_key <- name
      metadata_key <- gsub("Kellis", "MIT", metadata_key)      
      metadata_key <- gsub("DeJager", "CUIMC1", metadata_key)  
      
      match_row <- contexts_metadata[contexts_metadata$context == metadata_key, ]
      
        if(nrow(match_row) > 0) {
        result[i] <- match_row$context[1]
      } else {
        result[i] <- name
      }
    } else {
      match_row <- contexts_metadata[contexts_metadata$context == name, ]
      
      if(nrow(match_row) > 0) {
        result[i] <- match_row$context[1]
      } else {
        result[i] <- name
      }
    }
  }
  
  return(result)
}
```

```r
# Simplify: just use original resource names with spaces for display
detectable_data$display_resource <- gsub("_", " ", detectable_data$resource)

detectable_data$display_resource <- gsub("Kellis", "MIT", detectable_data$display_resource)
detectable_data$display_resource <- gsub("DeJager", "CUIMC1", detectable_data$display_resource)

# Fix all patterns to be Cohort Context Modality format:

# 1. STARNET - change "STARNET eQTL Mac" to "STARNET Mac eQTL"
detectable_data$display_resource <- gsub("STARNET eQTL Mac", "STARNET Mac eQTL", detectable_data$display_resource)

# 2. MSBB eQTL - change "BM 44 MSBB eQTL" to "MSBB BM44 eQTL" etc.
detectable_data$display_resource <- gsub("BM (\\d+) MSBB eQTL", "MSBB BM\\1 eQTL", detectable_data$display_resource)

# 3. MSBB pQTL - change "MSBB BM36 pQTL" to "MSBB BM36 pQTL" (already correct)

# 4. Bennett pQTL - change "DLPFC Bennett pQTL" to "Bennett DLPFC pQTL"
detectable_data$display_resource <- gsub("DLPFC Bennett pQTL", "Bennett DLPFC pQTL", detectable_data$display_resource)

# 5. Klein gpQTL - change "DLPFC Klein gpQTL" to "Klein DLPFC gpQTL"
detectable_data$display_resource <- gsub("DLPFC Klein (gpQTL adjusted|gpQTL unadjusted)", "Klein DLPFC \\1", detectable_data$display_resource)

# 6. monocyte - change "monocyte ROSMAP eQTL" to "ROSMAP monocyte eQTL"
detectable_data$display_resource <- gsub("monocyte ROSMAP eQTL", "ROSMAP monocyte eQTL", detectable_data$display_resource)

# 7. Knight pQTL - change "Knight pQTL brain" to "Knight brain pQTL"
detectable_data$display_resource <- gsub("Knight (eQTL|pQTL) brain", "Knight brain \\1", detectable_data$display_resource)

# 8. CUIMC1 cell types - change "Ast CUIMC1 eQTL" to "CUIMC1 Ast eQTL" etc.
detectable_data$display_resource <- gsub("(Ast|Oli|Mic|OPC|Exc|Inh|AC|PCC|DLPFC) CUIMC1 eQTL", "CUIMC1 \\1 eQTL", detectable_data$display_resource)

detectable_data$display_resource <- gsub("(Ast|Oli|Mic|OPC|Exc|Inh) MIT eQTL", "MIT \\1 eQTL", detectable_data$display_resource)
detectable_data$display_resource <- gsub("(Ast|Mic) (\\d+) MIT eQTL", "MIT \\1\\2 eQTL", detectable_data$display_resource)
# Change "Ast mega eQTL" to "mega Ast eQTL" etc.
detectable_data$display_resource <- gsub("(Ast|Oli|Mic|OPC|Exc|Inh) mega eQTL", "mega \\1 eQTL", detectable_data$display_resource)

# Calculate percentages and prepare total counts WITH PROPER AGGREGATION
detectable_data <- detectable_data %>%
  mutate(n_detectable_category = as.numeric(as.character(n_detectable_category))) %>%
  mutate(
    n_detectable_category = case_when(
      n_detectable_category == 0 ~ "pip sum ≥ 0.95",
      n_detectable_category == 1 ~ "1",
      n_detectable_category >= 2 & n_detectable_category <= 5 ~ "2-5",
      n_detectable_category >= 6 & n_detectable_category <= 9 ~ "6-9",
      n_detectable_category >= 10 ~ "≥10",
      TRUE ~ as.character(n_detectable_category)
    )
  ) %>%
  group_by(resource, display_resource, n_detectable_category) %>%
  summarize(
    gene_count = sum(gene_count),
    .groups = "drop"
  ) %>%
  group_by(resource) %>%
  mutate(
    total_count = sum(gene_count),
    percentage = round((gene_count / total_count) * 100, 1),
    log_gene_count = log10(gene_count)
  ) %>%
  ungroup()

# Set factor levels for LEFT-TO-RIGHT stacking
detectable_data <- detectable_data %>%
  mutate(n_detectable_category = factor(
    n_detectable_category,
    levels = c("≥10", "6-9", "2-5", "1", "pip sum ≥ 0.95")
  ))

# Add tissue grouping for faceting
detectable_data <- detectable_data %>%
  mutate(
    tissue_group = case_when(
      grepl("STARNET|ROSMAP|PCC|DLPFC|MSBB|Knight.*eQTL.*brain|AC.*CUIMC1|BM.*MSBB|PCC.*CUIMC1", display_resource) ~ "Bulk Brain Tissue",
      grepl("Inh|Exc", display_resource) ~ "Neurons",
      grepl("Knight.*pQTL|DLPFC.*Klein", display_resource) ~ "Other", 
      grepl("MIGA|MiGA", display_resource) ~ "MiGA",
      grepl("Metabrain", display_resource) ~ "MetaBrain",
      TRUE ~ "Glia"
    )
  )
# Create ordering within each tissue group
detectable_data <- detectable_data %>%
  group_by(tissue_group) %>%
  arrange(desc(total_count), .by_group = TRUE) %>%
  mutate(rank_in_group = row_number()) %>%
  ungroup()

# Define the order of tissue groups for faceting
tissue_group_order <- c("Bulk Brain Tissue", "Neurons", "Glia", "Other",
                        "MiGA", "MetaBrain")

detectable_data$tissue_group <- factor(detectable_data$tissue_group, levels = tissue_group_order)

# Calculate bar end positions in LOG space
bar_end_positions <- detectable_data %>%
  group_by(resource) %>%
  summarise(
    display_resource = first(display_resource),
    tissue_group = first(tissue_group),
    rank_in_group = first(rank_in_group),
    total_count = first(total_count),
    x_end = sum(log_gene_count),
    .groups = "drop"
  )

# Identify small segments (less than 5% or narrow in log space)
detectable_data <- detectable_data %>%
  mutate(
    is_small = percentage < 5 | log_gene_count < 0.15,  # Adjust threshold as needed
    label_inside = ifelse(is_small, "", paste0(percentage, "%")),
    label_outside = ifelse(is_small, paste0(percentage, "%"), "")
  )

# Calculate position for outside labels (at the middle of each segment in log space)
detectable_data <- detectable_data %>%
  group_by(resource) %>%
  arrange(desc(n_detectable_category)) %>%
  mutate(
    cumsum_log = cumsum(log_gene_count),
    x_middle = cumsum_log - log_gene_count/2
  ) %>%
  ungroup()

```
